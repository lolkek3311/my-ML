```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/lolkek3310/python/Подготовка Гомель/KMeans Clustering 2D Dataset/dataset.csv')

plt.figure(figsize=(10,6))
plt.scatter(df['x'],df['y'])
plt.show()
```

```python
X = df
X
```

```python
clustering_score = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    clustering_score.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1,11), clustering_score)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)

pred = kmeans.fit_predict(X)


pred
```

```python
df['Cluster'] = pred
df
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['x'], df['y'], c = df['Cluster'],
                      cmap = 'viridis', alpha = 0.6, s = 50)

plt.show()
```

```python
# Если хотите показать центроиды на графике
centroids = df.groupby('Cluster')[['x', 'y']].mean()

plt.figure(figsize=(10, 6))
plt.scatter(df['x'], df['y'], c=df['Cluster'])
plt.scatter(centroids['x'], centroids['y'], c='red', s=200, 
            marker='X', edgecolors='black', linewidths=2, label='Центроиды')
plt.show()
```
