## Кластеризация: Пример с синтетическими данными

### Что делает этот блок:
**Импортирует необходимые библиотеки и генерирует синтетические данные для демонстрации кластеризации.**

### Подробное объяснение:
- **`make_blobs`** — функция из `sklearn.datasets`, которая создаёт искусственные (синтетические) данные с заранее известным количеством кластеров. Это идеально подходит для обучения и демонстрации алгоритмов кластеризации.
- **`n_samples=300`** — создаём 300 точек данных.
- **`centers=4`** — данные будут сгруппированы вокруг 4 центров (кластеров).
- **`cluster_std`** — стандартное отклонение кластеров, определяет насколько "разбросаны" точки вокруг центров. Разные значения `[0.6, 0.8, 0.4, 0.7]` делают кластеры разной плотности.
- **`random_state=42`** — фиксирует генератор случайных чисел, чтобы при каждом запуске получались одни и те же данные (воспроизводимость результата).

### Зачем это нужно:
Синтетические данные позволяют наглядно увидеть, как алгоритм KMeans разделяет точки на группы, и сравнить результат с "правильным" ответом (так как мы знаем истинные метки кластеров).

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Генерируем данные
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 1. K-Means
kmeans = KMeans(n_clusters=4)
labels_km = kmeans.fit_predict(X)

# 2. DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels_db = dbscan.fit_predict(X)

# Визуализация
plt.scatter(X[:, 0], X[:, 1], c=labels_km, cmap='viridis')
plt.title("K-Means Clustering")
plt.show()
```
